```python
import math
import nltk
# при втором запуске программы строку с download можно убрать,
# чтобы он каждый раз не скачивал то, что уже есть
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from sklearn.externals import joblib



# метод определения класса
def classify(classifier, features):
    classes, freq = classifier
    return min(classes.keys(), key=lambda cls: -math.log(classes[cls]) + sum(
        -math.log(freq.get((cls, feat), 10 ** (-7))) for feat in features))


# метод предобработки текста
def preprocess(text, stemmer):
    # берем только слова и числа (\w+)
    tokenizer = nltk.RegexpTokenizer(r'\w+')
    words = tokenizer.tokenize(text.lower())
    # исключаем стоп-слова (союзы и т.д.) и выделяем основы оставшихся слов
    clean_words = [stemmer.stem(w) for w in words if not w in stopwords.words('russian')]
    return clean_words


stemmer = SnowballStemmer("russian")

clf = joblib.load('data/dicts.pkl')

result = classify(clf, preprocess("Последний день распродажи! Купи две шубы по цене одной, трать бонусы и получай подарки!", stemmer))
print("Спам" if result == 0 else "Не спам")
print(clf)
```
